```python
import pandas as pd
import numpy as np

ENDERECO_DADOS = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'

try:
    # Lendo os dados do arquivo CSV
    df_roubos = pd.read_csv(ENDERECO_DADOS, encoding='iso-8859-1', sep=';')
    
    # Exibindo as primeiras linhas do DataFrame para verificar os dados
    print(df_roubos['regiao'].unique())
except Exception as e:
    print(f"Erro ao ler os dados: {e}")
```

```python
# Tratamento de Dados

try:
    df_roubos = df_roubos[['cisp', 'regiao', 'munic', 'roubo_veiculo']]

    # corrigindo erro de digitação
    correcao = df_roubos['regiao'].str.startswith('Grande Niter', na=False)
    df_roubos.loc[correcao, 'regiao'] = 'Grande Niterói'

    print(df_roubos['regiao'].unique())

except Exception as e:
    print(f"Erro ao tratar os dados: {e}")
```

```python
# Processando as informações

try:
    df_roubos = df_roubos.groupby(['cisp','regiao','munic'], as_index=False)['roubo_veiculo'].sum()
    display(df_roubos)

except Exception as e:
    print(f"Erro ao processar os dados: {e}")
```

```python
# Calculando a medidas: Array, Media, Mediana, Total e Quartil
try:
    array_roubos = df_roubos['roubo_veiculo']

    media = np.mean(array_roubos)
    mediana = np.median(array_roubos)
    total = np.sum(array_roubos)
    q1 = np.percentile(array_roubos, 25)
    q3 = np.percentile(array_roubos, 75)

    print(f"Media: {media}")
    print(f"Mediana: {mediana}")
    print(f"Total: {total}")
    print(f"Quartil 1: {q1}")
    print(f"Quartil 3: {q3}")


except Exception as e:
    print(f"Erro ao calcular as medidas: {e}")
```

```python
# Identificando os maiores e menores valores
try:
    # Maiores
    df_maiores = df_roubos[df_roubos['roubo_veiculo'] > q3].copy()

    # Menores
    df_menores = df_roubos[df_roubos['roubo_veiculo'] < q1].copy()

    # Adicionando a coluna de flags
    df_maiores['flags'] = 'mais'
    df_menores['flags'] = 'menos'

    # Concatenando os DataFrames
    df_roubos_flags = pd.concat([df_maiores, df_menores], ignore_index=True)
    display(df_roubos_flags)

except Exception as e:
    print(f"Erro ao identificar os maiores e menores valores: {e}")
```

```python
# Exportando os Dados
try:
    df_roubos_flags.to_csv('dados_tratados.csv', index=False)
    # exportando para Excel df_roubos_flags.to_excel('dados_tratados.xlsx', index=False) e instalar a biblioteca openpyxl: pip install openpyxl
    df_roubos_flags.to_excel('dados_tratados.xlsx', index=False)


except Exception as e:
    print(f"Erro ao exportar os dados: {e}")
```
